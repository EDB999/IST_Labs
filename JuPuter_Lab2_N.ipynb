```python
import pandas as pd

df_bookings = pd.read_csv('./2_bookings.csv', sep=';', engine='python')

bookings_head = df_bookings.head(7)
```

```python
df_bookings.rename(columns={"Is Canceled" : "Canceled", "Lead Time" : "Time"})
```

```python
df_canceling = df_bookings[df_bookings['Is Canceled'] == False]
df_canceling['Country'].value_counts()
```

```python
df_bookings.rename(columns={"Is Canceled" : "Canceled", "Lead Time" : "Time", "arrival full date" : "afd", "Arrival Date Year" : "ADY", "Arrival Date Month" : "ADM", "Arrival Date Week Number" :
                            "ADWN", "Arrival Date Day of Month" : "ADDoM", "Stays in Weekend nights" : "SiWn", "Stays in week nights" : "SiWkn",  "Adults" : "A", "Children" : "C",
                            "Babies" : "B", "Meal" : "M", "Country" : "Cntry"})

df_bookings.filter(like='stays total nights')

df_nights_city_hotel = (df_bookings.query("Hotel == 'City Hotel'")
                        .groupby('Hotel')
                        .agg({'stays total nights' : 'mean'}).
                        round(2))

df_nights_city_hotel
```

```python
df_bookings.rename(columns={"Is Canceled" : "Canceled", "Lead Time" : "Time", "arrival full date" : "afd", "Arrival Date Year" : "ADY", "Arrival Date Month" : "ADM", "Arrival Date Week Number" :
                            "ADWN", "Arrival Date Day of Month" : "ADDoM", "Stays in Weekend nights" : "SiWn", "Stays in week nights" : "SiWkn",  "Adults" : "A", "Children" : "C",
                            "Babies" : "B", "Meal" : "M", "Country" : "Cntry"})

df_bookings.filter(like='stays total nights')

df_nights_resort_hotel = (df_bookings.query("Hotel == 'Resort Hotel'")
                        .groupby('Hotel')
                        .agg({'stays total nights' : 'mean'}).
                        round(2))

df_nights_resort_hotel
```
