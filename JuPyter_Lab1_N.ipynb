```python
df_taxi_amount_pickups = df_taxi['pickups'].sum()
df_taxi_amount_pickups
df_taxi_sort_by_borough = df_taxi.groupby('borough', as_index=False).agg({'pickups' : 'sum'}).sort_values('pickups', ascending=False)
df_taxi_sort_by_borough
```

```python
df_taxi_min_pickups = df_taxi.groupby('borough').agg({'pickups' : 'sum'})
df_taxi_min_pickups.idxmin()
```

```python
df_taxi_hday = df_taxi.groupby(['borough', 'hday'], as_index=False).agg({'pickups': 'mean'})

hday = df_taxi_hday.query('hday == "Y"').rename(columns={'pickups': 'pickups_hday'})
not_hday = df_taxi_hday.query('hday == "N"').rename(columns={'pickups': 'pickups_not_hday'})

merged_df_taxi = pd.merge(not_hday[['borough', 'pickups_not_hday']], hday[['borough', 'pickups_hday']], on='borough')

result = merged_df_taxi.query('pickups_hday > pickups_not_hday')

print(result)
```

```python
pickups_by_mon_bor = df_taxi.groupby(['borough', 'pickup_month'], as_index=False).agg({'pickups' : 'sum'})

pickups_by_mon_borpickups_by_mon_bor = pickups_by_mon_bor.sort_values('pickups', ascending=False)

print(pickups_by_mon_borpickups_by_mon_bor)
```
