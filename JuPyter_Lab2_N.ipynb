```python
#13
df_bookings['total_kids'] = df_bookings["Children"] + df_bookings["Babies"]

df_bookings['has_kids'] = df_bookings['total_kids'] > 0

churn_rate_kids = (df_bookings[(df_bookings['has_kids'] == True) & (df_bookings['Is Canceled'] == True)].shape[0] /
                        df_bookings[df_bookings['has_kids'] == True].shape[0]) * 100

churn_rate_no_kids = (df_bookings[(df_bookings['has_kids'] == False) & (df_bookings['Is Canceled'] == True)].shape[0] /
                          df_bookings[df_bookings['has_kids'] == False].shape[0]) * 100

max_churn_rate = max(churn_rate_kids, churn_rate_no_kids)
min_churn_rate = min(churn_rate_kids, churn_rate_no_kids)
print(f"Наибольший процент оттока: {max_churn_rate:.2f}%")
print(f"Наименьший процент оттока: {min_churn_rate:.2f}%")

```
