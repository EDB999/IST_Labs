```python
#7
def read_n_agg(path):

    df_company = pd.read_csv(path, sep =';', engine='python')

    df_income_avg = df_company.query('income >= 0').groupby("company", as_index=True).agg({'income' : 'mean'})

    return df_income_avg

result = read_n_agg('./3_companies.csv')
print(result)    
```

```python
#8
import pandas as pd

df_taxi = pd.read_csv('./3_taxi_peru.csv', sep=';', engine='python', parse_dates=['start_at', 'end_at', 'arrived_at'])

df_taxi

```

```python
#9

df_overall_orders = df_taxi.query("end_state == 'drop off'").groupby("end_state").agg({"source" : "count"})


df_counter_orders_source = df_taxi.query("end_state == 'drop off'").groupby("source").size()

df_max_orders_source = df_counter_orders_source.idxmax()


df_max_orders_quantity = df_counter_orders_source.max()


df_procent_from_max_to_overall = df_max_orders_quantity / df_overall_orders * 100 

print(f"Большой всего сделанных, завершённых заказов с платформы: {df_max_orders_source}")
print(f"Больше всего заказов процентно: \n{df_procent_from_max_to_overall}")
```

```python
#10

%matplotlib inline

df_absolute_values = df_taxi.groupby("source").agg({"source" : "count"}).rename(columns={"source" : "quantity_orders"})

df_absolute_values

sns.barplot(x='source', y='quantity_orders', data=df_absolute_values)
```

```python
#11

count_plot_icon = sns.countplot(df_taxi['icon'])

count_plot_icon
```

```python
#12

end_order_by_source = sns.countplot(x='end_state', hue='source', data=df_taxi)

end_order_by_source
```

```python
#13

df_drivers_value_rating = df_taxi.groupby("driver_score", as_index=0).agg({"driver_score" : "count"})

df_drivers_score = df_taxi["driver_score"].value_counts()

df_drivers_amount_scores = df_drivers_score.sum()

df_drivers_amount_scores

df_drivers_percantage_scores = (df_drivers_value_rating / df_drivers_amount_scores).mul(100).round(2)

resetable_drivers_rating = df_drivers_percantage_scores.reset_index()

driver_score_counts = resetable_drivers_rating.rename(columns={"index" : "driver_score", "driver_score" : "percentage"})

driver_score_counts


```

```python
#14

ax = sns.barplot(x="driver_score", y="percentage", data=driver_score_counts, color='purple', alpha=0.5)
ax.set(xlabel='Driver score', ylabel='Percentage')
sns.despine()
plt.show()
```

```python
#15

df_value_riders_rating = df_taxi.groupby("rider_score", as_index=0).agg({"rider_score" : "count"})

# df_value_riders_rating

df_riders_score = df_taxi["rider_score"].value_counts()

df_amount_riders_score = df_riders_score.sum()

# df_amount_riders_score

df_percentage_riders_rating = (df_value_riders_rating / df_amount_riders_score).mul(100).round(2)

# df_percentage_riders_rating

resetable_riders_rating = df_percentage_riders_rating.reset_index()

riders_score_counts = resetable_riders_rating.rename(columns={"index" : "rider_score", "rider_score" : "percentage"})

riders_score_counts

ay = sns.barplot(x="rider_score", y="percentage", data=riders_score_counts, color="red", alpha=0.5)
ay.set(xlabel="Rider Score", ylabel="Percantage")
sns.despine()
plt.show
```
