```python
#2.1

df_games_of_the_year_named = df_games.loc[:, ['Year', 'Name']]

df_games_mode = df_games_of_the_year_named.mode()

df_games_mode
```

```python
#2.2

df_games_of_the_year = df_games.loc[:, ['Year']]

df_games_mediane = df_games_of_the_year.median()

df_games_mediane
```

```python
#2.3

df_games_of_the_year = df_games.loc[:, ['Year']]

df_games_quantile = df_games_of_the_year.quantile(q=0.1)

df_games_quantile
```

```python
#2.4 - last

df_games_counting_by_year = df_games.groupby("Year").agg({"Name" : "count"}).rename(columns={"Name" : "quantity_of_the_games"})

df_games_counting_by_year


plt.figure(figsize=(10, 9))
sns.barplot(x="quantity_of_the_games", y="Year", data=df_games_counting_by_year, color="purple")


```

```python
#3

df_games_counting_by_platform = df_games.groupby("Platform").agg({"Name" : "count"}).rename(columns={"Name" : "quantity_of_the_games"})

df_games_counting_by_platform

df_games_summary_by_publisher = df_games_counting_by_platform.sum()

df_games_summary_by_publisher

df_games_counting_by_platform["Occurrence_Percentage"] = ((df_games_counting_by_platform / df_games_summary_by_publisher) * 100).round(2)

df_games_counting_by_platform.sort_values(by="quantity_of_the_games", ascending=False)

df_games_filtered_by_occurrence = df_games_counting_by_platform.query("Occurrence_Percentage > 7")

df_games_filtered_by_occurrence









```

```python
#4

df_games_counting_by_publisher = df_games.groupby("Publisher").agg({"Name" : "count"}).rename(columns={"Name" : "quantity_of_the_games"})

df_top3 = df_games_counting_by_publisher.sort_values(by="quantity_of_the_games", ascending=False)

df_top3.head(3)
```

```python
#5.1

df_publisher_nintendo = df_games.query("Publisher == 'Nintendo'")

df_publisher_nintendo_two_columns_NA = df_publisher_nintendo[["NA_Sales"]]

df_nintendo_NA_mean = df_publisher_nintendo_two_columns_NA.mean().round(2)

df_nintendo_NA_median = df_publisher_nintendo_two_columns_NA.median()

df_nintendo_NA_std = df_publisher_nintendo_two_columns_NA.std().round(2)

print("Среднее арифметическое компании Nintendo в Северной Америке: \n" + f"{df_nintendo_NA_mean}\n")
print("Медиана компании Nintendo в Северной Америке: \n" + f"{df_nintendo_NA_median}\n")
print("Среднеквадратическое компании Nintendo в Северной Америке: \n" + f"{df_nintendo_NA_std}\n")


```

```python
#5.2

df_publisher_nintendo_two_columns_EU = df_publisher_nintendo[["EU_Sales"]]

df_nintendo_EU_mean = df_publisher_nintendo_two_columns_EU.mean().round(2)

df_nintendo_EU_median = df_publisher_nintendo_two_columns_EU.median()

df_nintendo_EU_std = df_publisher_nintendo_two_columns_EU.std().round(2)

print("Среднее арифметическое компании Nintendo в Европе: \n" + f"{df_nintendo_EU_mean}\n")
print("Медиана компании Nintendo в Европе: \n" + f"{df_nintendo_EU_median}\n")
print("Среднеквадратическое компании Nintendo в Европе: \n" + f"{df_nintendo_EU_std}\n")
```

```python
#5.3

df_publisher_nintendo_two_columns_JP = df_publisher_nintendo[["JP_Sales"]]

df_nintendo_JP_mean = df_publisher_nintendo_two_columns_JP.mean().round(2)

df_nintendo_JP_median = df_publisher_nintendo_two_columns_JP.median()

df_nintendo_JP_std = df_publisher_nintendo_two_columns_JP.std().round(2)

df_nintendo_JP_mode = df_publisher_nintendo_two_columns_JP.mode()

print("Среднее арифметическое компании Nintendo в Японии: \n" + f"{df_nintendo_JP_mean}\n")
print("Медиана компании Nintendo в Японии: \n" + f"{df_nintendo_JP_median}\n")
print("Среднеквадратическое компании Nintendo в Японии: \n" + f"{df_nintendo_JP_std}\n")
print("Мода компании Nintendo в Японии: \n" + f"{df_nintendo_JP_mode}\n")
```

```python
#5.4

df_publisher_nintendo_two_columns_Other = df_publisher_nintendo[["Other_Sales"]]

df_nintendo_Other_mean = df_publisher_nintendo_two_columns_Other.mean().round(2)

df_nintendo_Other_median = df_publisher_nintendo_two_columns_Other.median()

df_nintendo_Other_std = df_publisher_nintendo_two_columns_Other.std().round(2)

print("Среднее арифметическое компании Nintendo в Японии: \n" + f"{df_nintendo_Other_mean}\n")
print("Медиана компании Nintendo в Северной америке: \n" + f"{df_nintendo_Other_median}\n")
print("Среднеквадратическое компании Nintendo в Японии: \n" + f"{df_nintendo_Other_std}\n")
```

```python
#5.5

df_combined = pd.concat([df_nintendo_Other_median, df_nintendo_JP_median, df_nintendo_EU_median, df_nintendo_NA_median], axis=1)

df_max_median_from_4_regions = df_combined.max(axis=1).max()

print("Максимальная медиана среди всех регионов: " + f"{df_max_median_from_4_regions}")
```

```python
#6.1

df_publisher_nintendo_JPN = df_games.query("Publisher == 'Nintendo'")

df_publisher_nintendo_JPN_genres = df_publisher_nintendo[["Genre","JP_Sales"]]

df_publisher_nintendo_JPN_genres


plt.figure(figsize=(12,8))
sns.boxplot(x="Genre", y="JP_Sales", color="purple", data=df_publisher_nintendo_JPN_genres)
plt.xlabel('All Genres')
plt.ylabel('Japanese Sales')
plt.yscale('log')

```

```python
#6.1.1

plt.figure(figsize=(12,8))
ax = sns.swarmplot(x="Genre", y="JP_Sales", color="purple", data=df_publisher_nintendo_JPN_genres)
```

```python
#6.2

df_publisher_nintendo_JPN_genres_mean = df_games.query("Publisher == 'Nintendo'").groupby("Genre").agg({"JP_Sales" : 'mean'}).rename(columns={"JP_Sales" : "Japan_Mean"}).round(2)

df_publisher_nintendo_JPN_genres_mean

```

```python
#6.3

df_publisher_nintendo_JPN_genres_mode = df_games.query("Publisher == 'Nintendo'").groupby("Genre").agg({"JP_Sales" : lambda x: x.mode()[0]}).rename(columns={"JP_Sales" : "Japan_Mode"}).round(2)

df_publisher_nintendo_JPN_genres_mode
```

```python
#6.4

df_publisher_nintendo_JPN_genres_median = df_games.query("Publisher == 'Nintendo'").groupby("Genre").agg({"JP_Sales" : "median"}).rename(columns={"JP_Sales" : "Japan_Median"}).round(2)

df_publisher_nintendo_JPN_genres_median
```

```python
#7

df_publisher_nintendo = df_games.query("Publisher == 'Nintendo'")

df_platform_nintendo_for_5_genres = df_publisher_nintendo.query("Genre == 'Fighting' | Genre == 'Simulation' | Genre == 'Platform' | Genre == 'Racing' | Genre == 'Sports'")

df_nintendo_global_sales_by_years = df_platform_nintendo_for_5_genres[["Genre", "Year", "Global_Sales"]]

df_nintendo_genres_for_global_sales_by_years = df_nintendo_global_sales_by_years.query("Genre == 'Sports'").groupby('Year').agg({"Global_Sales" : "max"}).reset_index()

df_nintendo_genres_for_global_sales_by_years

plt.figure(figsize=(12, 9))
plt.plot(df_nintendo_genres_for_global_sales_by_years['Year'], df_nintendo_genres_for_global_sales_by_years['Global_Sales'], marker='o', color="purple")
plt.title('Global Sales of Sports Genre Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.grid()
plt.show()

threshold = 20
peaks = df_nintendo_genres_for_global_sales_by_years[df_nintendo_genres_for_global_sales_by_years['Global_Sales'] > threshold]

print(peaks[['Year', 'Global_Sales']])s
```
